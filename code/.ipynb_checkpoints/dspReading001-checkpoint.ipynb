```python
t= np.linspace(0,10,101) #sec
A= 1 # volts
f= 1 # Hz
ϕ= 0 # rad
xcos= A * np.cos(2*π*f*t + ϕ)
t, xcos
```

```python
pl.plot(t, xcos); pl.grid()
```

```python
xsin= A * np.sin(2*π*f*t + ϕ)
t, xsin
pl.plot(t,xsin); pl.grid()
```

```python
# 增加解析度
t= np.linspace(0,10,1001) #sec
xsin= A * np.sin(2*π*f*t + ϕ)
t, xsin
pl.plot(t,xsin); pl.grid()
```

```python
# 信號相加，(解析度、長度都要一致)

t= np.linspace(0,10,1001) #sec

xcos= A * np.cos(2*π*f*t + ϕ)
xsin= A * np.sin(2*π*f*t + ϕ)

sumSig= xcos + xsin

pl.plot(t,xcos, t, xsin, t, sumSig); pl.grid()
```

```python
# 詳細解說 訊號的 Amplitude (in volt), Frequency (in Hz), Phase (in rad)

t= np.linspace(0,10,1001) #sec

A1= 10
A2= 20

xcos1= A1 * np.cos(2*π*f*t + ϕ)
xcos2= A2 * np.cos(2*π*f*t + ϕ)

pl.plot(t,xcos1, t, xcos2); pl.grid()
```

```python
A= 10
f1= 1
f2= 2

xcos1= A * np.cos(2*π*f1*t + ϕ)
xcos2= A * np.cos(2*π*f2*t + ϕ)

pl.plot(t,xcos1, t, xcos2); pl.grid()
```

```python
A= 10
f= 1
ϕ1= 0 
ϕ2= (2*π) * 1/4 

# 1個週期 為 (2*π) rad, 
# ϕ2 - ϕ1 = (2*π) * 1/4 
# 代表 信號2 比 信號1 【超前】 1/4 個週期。

t= np.linspace(0,3, 101)
xcos1= A * np.cos(2*π*f*t + ϕ1)
xcos2= A * np.cos(2*π*f*t + ϕ2)

pl.plot(t,xcos1, t, xcos2); pl.grid()
```
