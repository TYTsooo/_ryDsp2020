```python
# generate a cosine wave

import numpy as np
import matplotlib.pyplot as pl
import IPython.display as dp

T=     10  # sec
Fs= 16000  # Hz

dt= 1/Fs
N=  T*Fs

t= np.arange(0, T, dt)

n= np.arange(0, N, 1)

t.size, t, \
n.size, n
```

```python
#
f0=   440  # Hz

π= np.pi

xt= np.cos(2 * π * f0 * t)
xn= np.cos(2 * π * f0 * n/Fs)

# test whether they are equal ?
# it seems that they are most equal, but not all.
# this is due to the numerical precision constraint of numpy
# this can be shown in showing their sum of absolute difference
#
(xt==xn),           \
all(xt==xn),        \
np.abs(xt-xn).sum()
```

```python
# plot them to see
# and play them to hear

pl.figure()
pl.plot(t, xt);pl.grid()
pl.figure()
pl.plot(n, xn);pl.grid()
```

```python
xt_audio= dp.Audio(data=xt, rate=Fs)
xt_audio
```

```python
xn_audio= dp.Audio(data=xn, rate=Fs)
xn_audio
```

```python
n.shape, xn.shape, t.shape, xt.shape
```

```python
# 觀察一小段
N1=100
pl.plot(n[0:N1],xn[0:N1],'o')
pl.xlabel('n (sample points)')
pl.grid()
```

```python
pl.plot(t[0:N1],xt[0:N1],'o-')
pl.xlabel('t (sec)')
pl.grid()
```

```python
#
# 快速傅立葉轉換FFT，求頻譜 Spectrum
#
fft_x=  np.fft.fft(xn)
spec_x= np.abs(fft_x)

fft_x.shape
f= np.arange(0,Fs,Fs/fft_x.size)
pl.plot(f, spec_x);pl.grid()
```

```python
# 看一小段 頻率區間就好

K=10000
pl.plot(f[0:K], spec_x[0:K]);pl.grid()
```

```python
#
# Locate the frequency, where the amplitude of spectrum is max
# such a frequency is the dominant frequency component in the signal
#
kmax= np.argmax(spec_x) # this is the frequency index
fmax= f[kmax]           # this is frequency in Hz

kmax, fmax
```

# [Monte Carlo method](https://en.wikipedia.org/wiki/Monte_Carlo_method)

- 蒙地卡羅方法可用於近似計算圓周率：

讓電腦每次隨機生成兩個0到1之間的數，看以這兩個實數為橫縱坐標的點是否在單位圓內。生成一系列隨機點，統計單位圓內的點數與總點數，（圓面積和正方形面積之比為PI:4，PI為圓周率），當隨機點取得越多時，其結果越接近於圓周率

![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/84/Pi_30K.gif/330px-Pi_30K.gif)
    

 - 一次模擬

     - 每次產生 1,000,000 隨機的 [0,1] 之間的數，計算到原點之距離，小於1者計數。
     - 除以總數 1,000,000 之後，作為 撒下圓球會落入1/4圓內的機率p，該機率p = π/4
     - 因此 π 的估測值 = p * 4

     - 這個實驗做 N=1000次，取平均，
     - 考慮 標準差 std 及 樣本數=1000之後，
     - 計算 π 之 95% 信心區間


```python
#
# 每次產生 1,000,000 隨機的 [0,1] 之間的數，計算到原點之距離，小於1者計數。
# 除以總數 1,000,000 之後，作為 撒下圓球會落入1/4圓內的機率p，該機率p = π/4
# 因此 π 的估測值 = p * 4
#
# 這個實驗做 N=1000次，取平均，
# 考慮 標準差 std 及 樣本數=1000之後，
# 計算 π 之 95% 信心區間
#

N=1000
aL=[]
for i in range(N):
    np.random.seed(np.random.randint(0,N))

    X= np.random.random([2,1000000])
    pi0= ((X**2).sum(axis=0)< 1).sum()/X.shape[1]*4
    
    pi1= ((X**2).sum(axis=0)<=1).sum()/X.shape[1]*4
    #pi= pi1
    pi= (pi0+pi1)/2
    
    aL+=[pi]
ϵ= np.array(aL).std()/N**.5*2
μ= np.array(aL).mean()

μ-ϵ,μ,μ+ϵ 

# (3.141513420582042, 3.141630356, 3.141747291417958)
# (3.1415022503262167, 3.1416195880000006, 3.1417369256737846)
# pi0= (3.1415051140767334, 3.1416224240000004, 3.1417397339232673)
# pi1= (3.141509791975295,  3.1416270080000004, 3.141744224024706)
```

```python
np.pi>=μ-ϵ and np.pi<=μ+ϵ 
```

```python
x0= X[:,((X**2).sum(axis=0)< 1.0)]
x1= X[:,((X**2).sum(axis=0)> 1.0)]
x2= X[:,((X**2).sum(axis=0)== 1.0)]
```
